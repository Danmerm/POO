Classes-Encapsulamento1
Exemplo de código em Python que cria uma classe Pessoa com os atributos nome, idade e cpf, utilizando encapsulamento para proteger esses atributos. O encapsulamento é feito usando atributos privados (com __) e métodos getters e setters para acessar e modificar esses atributos de forma controlada.

Explicação do Código:
Atributos Privados:

Os atributos __nome, __idade e __cpf são privados (usando __), o que significa que não podem ser acessados diretamente fora da classe.

Getters e Setters:
Os métodos @property são usados para criar getters, que permitem acessar os valores dos atributos.
Os métodos @atributo.setter são usados para criar setters, que permitem modificar os valores dos atributos de forma controlada, com validações.

Validações:

No setter de nome, verifica-se se o novo valor é uma string não vazia.
No setter de idade, verifica-se se o novo valor é um inteiro positivo.
No setter de cpf, verifica-se se o novo valor é uma string com 11 caracteres.
Método exibir_informacoes:
Um método para exibir todas as informações da pessoa de forma organizada.

Benefícios do Encapsulamento:
Proteção dos dados: Atributos só podem ser modificados de forma controlada, evitando valores inválidos.
Flexibilidade: Você pode alterar a implementação interna da classe sem afetar o código que a utiliza.
Manutenção: Facilita a manutenção do código, pois as regras de validação estão centralizadas nos setters.
Esse exemplo é um ótimo ponto de partida para entender encapsulamento em POO!

Classes-Polimorfismo1
Exemplo de código em Python que demonstra o uso de herança, polimorfismo e sobrescrita de métodos. A classe base Animal define um método fazer_som(), e as classes derivadas Cachorro, Gato e Pato sobrescrevem esse método para retornar sons específicos. O polimorfismo é demonstrado ao tratar objetos de diferentes classes de forma uniforme.
Explicação do Código:

Classe Base Animal:
Define um método fazer_som() que levanta uma exceção NotImplementedError. Isso força as classes derivadas a implementar esse método.

Classes Derivadas:
Cachorro, Gato e Pato herdam da classe Animal e sobrescrevem o método fazer_som() para retornar sons específicos.

Polimorfismo:
A função ouvir_som(animal) recebe um objeto do tipo Animal (ou qualquer classe derivada) e chama o método fazer_som(). O método específico da classe derivada é executado, demonstrando polimorfismo.

Lista de Animais:
Objetos de diferentes classes derivadas são armazenados em uma lista e tratados de forma uniforme, chamando o método fazer_som() para cada um.

Benefícios do Polimorfismo:
Flexibilidade: Permite tratar objetos de diferentes classes de forma uniforme, desde que compartilhem uma interface comum (neste caso, o método fazer_som()).
Extensibilidade: Novas classes derivadas podem ser adicionadas sem modificar o código existente que usa polimorfismo.
Organização: O código fica mais modular e fácil de manter.
Esse exemplo é uma ótima introdução ao conceito de polimorfismo em POO

O método __str__ é sobrescrito para fornecer uma representação legível do autor.

Classe Livro:
Representa um livro com atributos titulo, ano_publicacao e autor.
O atributo autor é uma instância da classe Autor, demonstrando composição.
O método __str__ é sobrescrito para fornecer uma representação legível do livro.

Classe Biblioteca:
Gerencia uma coleção de livros usando uma lista.

Métodos:
adicionar_livro: Adiciona um livro à biblioteca.
listar_livros: Lista todos os livros na biblioteca.
buscar_livros_por_autor: Busca livros de um autor específico.
Composição:
A classe Livro contém uma instância da classe Autor, demonstrando que um livro tem um autor.

Exemplo de Uso:
Cria autores, livros e uma biblioteca.

Classes-Abstraçao
Exemplo de código em Python que utiliza o módulo abc para criar uma classe abstrata Forma com um método abstrato calcular_area(). Em seguida, são implementadas classes derivadas como Retangulo, Circulo e Triangulo, que herdam de Forma e implementam o método calcular_area().

Explicação do Código:
Classe Abstrata Forma:
Define um método abstrato calcular_area() usando o decorador @abstractmethod.

Classes derivadas devem implementar esse método, caso contrário, um erro será levantado.
Classes Derivadas:
Retangulo: Calcula a área como base * altura.
Circulo: Calcula a área como π * raio².
Triangulo: Calcula a área como (base * altura) / 2.
RegAnularCircular: Calcula a área como π * (raio_2²-raio_1²).
Trapezio: Calcula a área como (base_1+base_2)* altura / 2.

Polimorfismo:
A função exibir_area(forma) recebe um objeto do tipo Forma (ou qualquer classe derivada) e chama o método calcular_area(). O método específico da classe derivada é executado, demonstrando polimorfismo.

Lista de Formas:
Objetos de diferentes classes derivadas são armazenados em uma lista e tratados de forma uniforme, chamando o método calcular_area() para cada um.

Benefícios do Uso de Classes Abstratas:
Força a implementação: Garante que todas as subclasses implementem o método calcular_area().
Organização: Define uma interface comum para todas as formas, facilitando a extensão do código.
Polimorfismo: Permite tratar objetos de diferentes classes de forma uniforme.

O titular deve ser uma string não vazia.
Depósitos e saques só são permitidos com valores positivos.

Classes-Observer
um exemplo de código em Python que implementa o padrão Observer. A classe Sujeito mantém uma lista de Observadores e os notifica sempre que seu estado muda. Os observadores são notificados automaticamente e podem reagir às mudanças de estado.

Explicação do Código:
Classe Observador:
Define uma interface com o método abstrato atualizar(mensagem), que deve ser implementado por todas as classes concretas de observadores.

Classe Sujeito:
Mantém uma lista de observadores (__observadores).
Oferece métodos para adicionar (adicionar_observador), remover (remover_observador) e notificar (notificar_observadores) os observadores.
Quando o estado do sujeito muda (alterar_estado), todos os observadores são notificados.

Classe ObservadorConcreto:
Implementa a interface Observador e define como o observador reage às notificações (neste caso, exibindo uma mensagem).

Funcionamento:
Os observadores são registrados no sujeito.
Quando o estado do sujeito muda, ele notifica todos os observadores registrados.
Os observadores reagem às notificações de acordo com sua implementação.

Benefícios do Padrão Observer:
Desacoplamento: O sujeito não precisa conhecer detalhes dos observadores, apenas que eles implementam a interface Observador.
Extensibilidade: Novos observadores podem ser adicionados sem modificar o sujeito.
Reatividade: Os observadores reagem automaticamente às mudanças de estado do sujeito.

Benefícios do Encapsulamento:
Proteção dos dados: Atributos privados só podem ser acessados ou modificados de forma controlada.
Validações: Garante que o saldo nunca seja negativo e que operações sejam realizadas apenas com valores válidos.
Flexibilidade: Permite alterar a implementação interna da classe sem afetar o código que a utiliza.

Classes-Polimorfismo3
um exemplo de código em Python que implementa um jogo simples onde dois personagens (Guerreiro e Mago) lutam entre si. Cada personagem tem atributos como vida, ataque e defesa, e métodos como atacar() e receber_dano(). O polimorfismo é utilizado para implementar habilidades especiais de cada personagem.

Explicação do Código:
Classe Base Personagem:
Define atributos comuns (nome, vida, ataque, defesa) e métodos abstratos (atacar e receber_dano).
O método esta_vivo verifica se o personagem ainda tem vida.

Classe Guerreiro:
Implementa atacar para causar dano ao inimigo.
Implementa receber_dano para reduzir a vida do guerreiro.
Possui uma habilidade especial (habilidade_especial) que aumenta seu ataque.

Classe Mago:
Implementa atacar para causar dano ao inimigo.
Implementa receber_dano para reduzir a vida do mago.
Possui uma habilidade especial (habilidade_especial) que restaura sua vida.

Polimorfismo:
A função batalha trata os personagens de forma uniforme, chamando os métodos atacar, receber_dano e habilidade_especial sem precisar saber o tipo específico de cada personagem.

Batalha:
A batalha ocorre em turnos, com cada personagem atacando o outro.
A cada 3 turnos, os personagens usam suas habilidades especiais.
A batalha termina quando um dos personagens perde toda a vida.

Benefícios do Polimorfismo:
Flexibilidade: Permite adicionar novos tipos de personagens sem modificar a lógica da batalha.
Extensibilidade: Novas habilidades ou comportamentos podem ser facilmente adicionados.
Organização: O código fica mais modular e fácil de manter.
